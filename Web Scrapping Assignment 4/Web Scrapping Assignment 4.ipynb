Question 6. Scrape the details of Highest selling novels.  
A) Book name  
B) Author name  
C) Volumes sold  
D) Publisher  
E) Genre  
 Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare   


```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

books = soup.find_all('tr', {'class': 'item'})

with open('books.csv', mode='w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Book Name', 'Author Name', 'Volumes Sold'])

    for book in books:
        title = book.find('a').text.strip()
        author = book.find('span', {'class': 'author'}).text.strip()
        sold = book.find('td', {'class': 'sales'}).text.strip()

        writer.writerow([title, author, sold])
```

Question 8: Details of Datasets from UCI machine learning repositories.  
Url = https://archive.ics.uci.edu/ You  
have to find the following details:  
A) Dataset name  
B) Data type  
C) Task  
D) Attribute type  
E) No of instances  
F) No of attribute G) Year  


```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://archive.ics.uci.edu/ml/datasets.php'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

datasets = soup.find_all('tr', {'class': 'data'})

with open('datasets.csv', mode='w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Dataset Name', 'Data Type', 'Task', 'Attribute Type', 'No of Instances', 'No of Attributes', 'Year'])

    for dataset in datasets:
        name = dataset.find('a').text.strip()
        link = dataset.find('a')['href']
        url = f'https://archive.ics.uci.edu/ml/{link}'
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        data_type = soup.find('td', {'class': 'dataType'}).text.strip()
        task = soup.find('td', {'class': 'task'}).text.strip()
        attribute_type = soup.find('td', {'class': 'attributeType'}).text.strip()
        no_of_instances = soup.find('td', {'class': 'instances'}).text.strip()
        no_of_attributes = soup.find('td', {'class': 'attributes'}).text.strip()
        year = soup.find('td', {'class': 'year'}).text.strip()

        writer.writerow([name, data_type, task, attribute_type, no_of_instances, no_of_attributes, year])
```

Question 5: Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/ You have to find the  following details:  
A) Song name  
B) Artist name  
C) Last week rank  
D) Peak rank  
E) Weeks on board  


```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://www.billboard.com/charts/hot-100/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

songs = soup.find_all('tr', {'class': 'o-chart-results-list-row'})

with open('billboard_hot_100.csv', mode='w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Song Name', 'Artist Name', 'Last Week Rank', 'Peak Rank', 'Weeks on Board'])

    for song in songs:
        name = song.find('h3', {'class': 'o-chart-results-list-row__title'}).text.strip()
        artist = song.find('span', {'class': 'c-label a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only'}).text.strip()
        last_week_rank = song.find('span', {'class': 'o-chart-results-list-row__rank'}).text.strip()
        peak_rank = song.find('span', {'class': 'o-chart-results-list-row__rank u-letter-spacing-0021'}).text.strip()
        weeks_on_board = song.find('span', {'class': 'o-chart-results-list-row__position'}).text.strip()

        writer.writerow([name, artist, last_week_rank, peak_rank, weeks_on_board])
```

Question 4: Scrape the details of trending repositories on Github.com.  
Url = https://github.com/  
You have to find the following details:  
A) Repository title  
B) Repository description  
C) Contributors count  
D) Language used 


```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'https://github.com/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

repos = soup.find_all('article', {'class': 'Box-row'})

with open('github_trending.csv', mode='w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(['Repository Title', 'Repository Description', 'Contributors Count', 'Language Used'])

    for repo in repos:
        title = repo.find('h1', {'class': 'h3 lh-condensed'}).text.strip()
        description = repo.find('p', {'class': 'col-9 text-gray my-1 pr-4'}).text.strip()
        contributors = repo.find('a', {'href': '/' + title.lower().replace(' ', '-') + '/graphs/contributors'}).text.strip()
        language = repo.find('span', {'itemprop': 'programmingLanguage'})
        if language:
            language = language.text.strip()
        else:
            language = ''

        writer.writerow([title, description, contributors, language])
```
